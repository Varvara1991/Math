```python
#Тема 2. Задание 3.1
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math

x=[]
y1=[]
y2=[]
x0=0
y0=0
r=2
for i in range (100*2*r+1):
    x_=-r+i/100
    x.append(x_)
    y1.append((r**2-(x_-x0)**2)**(1/2)+y0)
    y2.append(((r**2-(x_-x0)**2)**(1/2)+y0)*(-1))
plt.plot(x,y1, color='blue')
plt.plot(x,y2, color='blue')
plt.ylim(-3,3) 
plt.xlim(-4.5,4.5)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
#Тема 3. Задание 4.1
from scipy.optimize import fsolve

#y = x*2 – 1
#np.exp(x) + x*(1 – y) = 1
#x*(1 – y) = 1 - np.exp(x)
#1 – y = (1 - np.exp(x))/x
#y = 1 - ((1 - np.exp(x))/x)

x = np.linspace(-3*np.pi, 3*np.pi)
plt.plot(x,(x**2-1))
plt.plot(x,1 - ((1 - np.exp(x))/x))
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-5,5)
plt.ylim(-1,10) 
plt.grid(True)
plt.show()

def equations(p):
    x,y = p
    return(y - x**2 + 1, np.exp(x) + x*(1 - y) - 1)
x1,y1 = fsolve(equations,(-2,4))
print(x1,y1)

x2,y2 = fsolve(equations,(2,6))
print(x2,y2)
```

```python
#Тема 3. Задание 4.2
# y = x**2 – 1
# np.exp(x) + x*(1 – y) > 1
# np.exp(x)+x*(2−x**2)−1>0

x = np.linspace(-3*np.pi, 3*np.pi, 1000)
plt.plot(x,np.exp(x)+x*(2-x**2)-1)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-5,5)
plt.ylim(-10,10) 
plt.grid(True)
plt.show()

def equations(x):
    return(np.exp(x)+x*(2-x**2)-1)

x1 = fsolve(equations,(-2))
print(x1)

x2 = fsolve(equations,(0))
print(x2)

x3 = fsolve(equations,(2))
print(x3)

x4 = fsolve(equations,(4))
print(x4)

print('Ответ: x<-1.58, 0<x<2.618, x>4.2')
```
